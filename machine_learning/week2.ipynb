```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Get some data:

```python
from sklearn.datasets import fetch_california_housing

temp_housing = fetch_california_housing()
data = np.c_[temp_housing.data, temp_housing.target]
columns = np.append(temp_housing.feature_names, ['MedianHouseValue'])
housing = pd.DataFrame(data, columns=columns)

housing.head()
```

Look at some stats if you'd like!

```python
#@title
housing.describe()
```

```python
Represent the features:
```

```python
housing.hist(bins=50, figsize=(10, 10))
plt.show()
```

Plot longitude and latitude:

```python
housing.plot(kind='scatter', x='Longitude', y='Latitude')
```

Interesting, but could be more informative. The parameter alpha controls the opacity of the scatterplot: by decreasing alpha, higher density areas will show up darker.

```python
housing.plot(kind='scatter', x='Longitude', y='Latitude', alpha=.2)
```

```python
Even more features can be combined into scatter plots!
```

```python
housing.plot(
    kind='scatter', 
    x='Longitude', 
    y='Latitude', 
    alpha=.2, 
    s=housing['Population']/100, 
    label='population', 
    c='MedianHouseValue', 
    cmap=plt.get_cmap('jet'), 
    colorbar=True
  )
```
